```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/Logistics_Cleaned_Dataset.csv')
df
```

```python
1.	Heatmap → Correlation matrix.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = df[['weight_kg', 'distance_km', 'cost',
           'selling_price', 'delivery_time_days']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
2.	Boxplot → Transport Mode vs Delivery Time.
```

```python
x_col = df.columns[df.columns.str.contains('transport', case=False)][0]
y_col = df.columns[df.columns.str.contains('delivery', case=False)][0]
sns.boxplot(x=x_col, y=y_col, data=df)
plt.show()
```

```python
3.	Countplot → Vehicle Types.
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x='vehicle_type',
    data=df
)
plt.title("Count of Shipments by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Shipments")
plt.xticks(rotation=45)
plt.show()
```

```python
4.	Violin → Cost vs Transport Mode.
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(
    x='transport_mode',
    y='cost',
    data=df
)
plt.title("Cost Distribution by Transport Mode")
plt.xlabel("Transport Mode")
plt.ylabel("Cost")
plt.xticks(rotation=45)
plt.show()
```

```python
5.	Scatterplot → Weight vs Fuel Consumption.
```

```python
fuel_col = df.columns[df.columns.str.contains('fuel', case=False)][0]
sns.scatterplot(
    x='weight_kg',
    y=fuel_col,
    data=df
)
plt.title("Weight vs Fuel Consumption")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df['profit'] = df['selling_price'] - df['cost']
city_col = df.columns[df.columns.str.contains('city')][0]
plt.figure(figsize=(10,5))
sns.barplot(
    x=city_col,
    y='profit',
    data=df
)

plt.title("City vs Profit")
plt.xlabel("City")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.show()
```
