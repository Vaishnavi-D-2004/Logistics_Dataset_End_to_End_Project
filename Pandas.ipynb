```python
import pandas as pd
import numpy as np
df=pd.read_csv('/logistics_uncleaned_5000rows.csv')
df
```

```python
1.	Load data, show info(), describe().
```

```python
df.info()
```

```python
df.describe()
```

```python
2.Identify Missing Values.
```

```python
df.isnull().sum()
```

```python
3.Replace "NULL", "NaN", "" with np.nan.
```

```python
df.replace(["NULL", "NaN", ""], np.nan, inplace=True)
df
```

```python
4.Fix negative values in:Weight_kg,Cost,Distance_km
```

```python
import numpy as np
import pandas as pd
cols = ['Weight_Kg', 'Cost', 'Distance_Km']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df[cols] = df[cols].mask(df[cols] < 0, np.nan)
df
```

```python
5.Convert Shipment_Date & Delivery_Date to datetime.
```

```python
df['Shipment_Date'] = pd.to_datetime(df['Shipment_Date'], dayfirst=True, errors='coerce')
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'], dayfirst=True, errors='coerce')
df
```

```python
6.Create Delivery_Time_Calc = Delivery_Date â€“ Shipment_Date.
```

```python
df['Delivery_Time_Calc'] = (df['Delivery_Date'] - df['Shipment_Date']).dt.days
df
```

```python
7.Validate Delivery_Time_Days column.
```

```python
df['Delivery_Time_Diff'] = df['Delivery_Time_Days'] - df['Delivery_Time_Calc']
df
```

```python
8.Group by City â†’ Average Delivery Time.
```

```python
avg_delivery_city = (
    df.groupby('Destination_City')['Delivery_Time_Calc']
      .mean()
      .reset_index()
)
avg_delivery_city
```

```python
9.Group by Transport_Mode â†’ Total Shipments
```

```python
shipments_by_mode = (
    df.groupby('Transport_Mode')
      .size()
      .reset_index(name='Total_Shipments')
)
shipments_by_mode
```

```python
10.Find top 10 profitable shipments.
```

```python
df['Profit'] = df['Selling_Price'] - df['Cost']
df
```

```python
top10_profitable = df.sort_values(by='Profit', ascending=False).head(10)
top10_profitable
```

```python
11.	Generate summary per vendor.
```

```python
vendor_summary = (
    df.groupby('Vendor')
      .agg(
          Total_Shipments=('Shipment_ID', 'count'),
          Total_Revenue=('Selling_Price', 'sum'),
          Total_Cost=('Cost', 'sum'),
          Total_Profit=('Profit', 'sum')
      )
      .reset_index()
)
vendor_summary
```

```python
12.	Export cleaned dataset â†’ cleaned_logistics.csv.
```

```python
df.to_csv('cleaned_logistics.csv', index=False)
df
```
