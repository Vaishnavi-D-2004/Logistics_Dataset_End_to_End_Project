```python
2.	Find outlier limits using IQR for Weight.
```

```python
df.columns = df.columns.str.strip().str.lower()
df
```

```python
weight_col = df.select_dtypes(include='number').columns[0]
weight_col
```

```python
q1 = df[weight_col].quantile(0.25)
q3 = df[weight_col].quantile(0.75)
iqr = q3 - q1

lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

lower_limit, upper_limit
```

```python
Probability
1.Poisson: Expected number of delays per week.
```

```python
import numpy as np
from scipy.stats import poisson, binom, norm
```

```python
lambda_delays = 5   # average delays per week
k = 3               # number of delays
```

```python
poisson_prob = poisson.pmf(k, lambda_delays)
poisson_prob
```

```python
2.Binomial: Probability of 3 damaged shipments out of 20
```

```python
from scipy.stats import binom
```

```python
n = 20     # total number of shipments
k = 3      # number of damaged shipments
p = 0.1    # probability of damage (10%)
```

```python
binomial_prob = binom.pmf(k, n, p)
binomial_prob
```
