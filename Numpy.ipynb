```python
import numpy as np

df['Profit_Margin'] = np.where(
    df['Selling_Price'] > 0,
    ((df['Selling_Price'] - df['Cost']) / df['Selling_Price']) * 100,
    np.nan
)
df
```

```python
3.	Identify extreme outliers using NumPy percentile.
```

```python
import numpy as np

# Convert Cost column to NumPy array
cost = df['Cost'].to_numpy()

# Calculate Q1, Q3, IQR
Q1 = np.nanpercentile(cost, 25)
Q3 = np.nanpercentile(cost, 75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = cost[(cost < lower_bound) | (cost > upper_bound)]

lower_bound, upper_bound, outliers[:10]
```

```python
4.	Simulate 1000 random delivery delays (Poisson).
```

```python
import numpy as np
# Set average delay (lambda)
avg_delay = 3   # 3 days average delay
# Simulate 1000 delivery delays
delivery_delays = np.random.poisson(lam=avg_delay, size=1000)
delivery_delays
```

```python
5.	Probability shipment is damaged (binomial).
```

```python

# Number of shipments
n_shipments = 1000

# Probability of damage
p_damage = 0.05   # 5% chance

# Simulate damaged shipments
damaged_shipments = np.random.binomial(
    n=n_shipments,
    p=p_damage,
    size=1
)
damaged_shipments
```
